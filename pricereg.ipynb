```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
```

```python
df_train.info()
```

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
```

```python
df_train
```

```python
df_train['running'] = df_train['running'].apply(lambda x: float(x.split()[0] if x.split()[1]=='km' else float(x.split()[0])*1.6))
```

```python
df_test['running'] = df_test['running'].apply(lambda x: float(x.split()[0] if x.split()[1]=='km' else float(x.split()[0])*1.6))
```

```python
df_train = df_train.drop_duplicates()
```

```python
df_concat = pd.concat([df_train,df_test])
```

```python
df_concat = df_concat.drop('Id',axis=1)
df_concat = df_concat.drop('wheel',axis=1)
```

```python
df_concat
```

```python
df_dummies = pd.get_dummies(df_concat, columns = ['model','motor_type','color','status','type'])
```

```python
df_dummies
```

```python
df_dummies.corr()['price']
```

```python
test = df_dummies[df_dummies['price'].isna()]
```

```python
train = df_dummies[~df_dummies['price'].isna()]
```

```python
test = test.drop('price',axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X = train.drop('price',axis=1)
y = train['price']
```

```python
X = scaler.fit_transform(X)
```

```python
test = scaler.transform(test)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from catboost import CatBoostRegressor
```

```python
model = CatBoostRegressor()
```

```python
param_grid = {
    'learning_rate': [0.01, 0.05, 0.1],
    'n_estimators': [100, 200, 300],
    'depth': [3, 5, 7],
    'l2_leaf_reg': [1, 3, 5, 7],
    'subsample': [0.8, 0.9, 1.0],
}
```

```python
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, verbose=2,scoring='neg_mean_squared_error')
```

```python
grid_search.fit(X,y)
```

```python
pred = grid_search.predict(test)
```

```python
grid_search.best_params_
```

```python
sub = pd.read_csv('sample_submission.csv')
```

```python
sub['price'] = pred
```

```python
sub.to_csv('submition.csv',index=False,index_label=False)
```
